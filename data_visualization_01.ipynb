```python
###HİSTOGRAM VE YOĞUNLUK GRAFİKLERİ
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
sns.displot(df.price, bins = 50, kde = True); 
#kde = yoğunluk argümanı
#bins = diyagramda istenilen sütun sayısı
#* ?sns.distplot
```

```python
###Çaprazlamalar
sns.kdeplot(df.price, fill = True); #fill or shade
```

```python
(sns
.FacetGrid(df,
            hue = "cut",
            height = 5,
            xlim = (0, 15000))
.map(sns.kdeplot, "price", fill = True)
.add_legend() #sağ tarafta oluşan cut başlıklı renkli kutucuklar için
);
```

```python
sns.catplot(x = "cut", y = "price", hue = "color", kind = "point", data = df);
```
