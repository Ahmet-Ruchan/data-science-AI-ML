```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset('diamonds')
df = diamonds.copy()
df.head()
```

```python
sns.displot(df.price, bins = 50, kde = True); 

#kde = density argument
#bins = number of columns requested in the diagram
#* ?sns.distplot
```

```python
sns.kdeplot(df.price, fill = True); #fill or shade
```

```python
(sns
.FacetGrid(df,
            hue = "cut",
            height = 5,
            xlim = (0, 15000))
.map(sns.kdeplot, "price", fill = True)
.add_legend() #for the colored boxes with cut caps on the right
);
```

```python
sns.catplot(x = "cut", y = "price", hue = "color", kind = "point", data = df);
```
